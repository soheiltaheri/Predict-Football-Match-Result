```python
years = list(range(2023,2020, -1))
```

```python
years
```

```python
all_matches =[]
```

```python
standing_url = 'https://fbref.com/en/comps/13/Ligue-1-Stats'
# standing_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
# standing_url = 'https://fbref.com/en/comps/12/La-Liga-Stats'
# standing_url = 'https://fbref.com/en/comps/11/Serie-A-Stats'
# standing_url = 'https://fbref.com/en/comps/20/Bundesliga-Stats'
```

```python
for year in years:
    data = requests.get(standing_url)
    soup = BeautifulSoup(data.text)
    standing_table = soup.select('table.stats_table')[0]

    links = [l.get('href') for l in standing_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    team_urls = [f'https://fbref.com{l}' for l in links]

    previous_season = soup.select('a.prev')[0].get('href')
    standing_url = f'https://fbref.com{previous_season}'

    for team_urls in team_urls:
        team_name = team_urls.split('/')[-1].replace('-Stats', '').replace('-', '')

        data =requests.get(team_urls)
        matches = pd.read_html(data.text, match='Scores & Fixtures')[0]

        soup = BeautifulSoup(data.text)
        links = [l.get('href') for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f'https://fbref.com{links[0]}')
        shooting = pd.read_html(data.text, match='Shooting')[0]
        shooting.columns = shooting.columns.droplevel()

        try:
            team_data = matches.merge(shooting[['Date', 'Sh', 'SoT', 'Dist', 'FK', 'PK', 'PKatt']], on='Date')
        except ValueError:
            continue

        team_data = team_data[team_data['Comp'] == 'La Liga']
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(1)
```

```python
 match_df = pd.concat(all_matches)
```

```python
match_df.to_csv('Data23-22-21-LaLiga.csv')
```
